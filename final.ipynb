```python
import os
```

```python
#importing the video
```

```python
cap = cv2.VideoCapture(r"C:\Users\dhananjayan\Downloads\stitch.avi") 
```

```python
#creating the folder names data_las
```

```python
try:
    if not os.path.exists("data_las"):
        os.makedirs("data_las")
except OSError: 
    print ('Error: Creating directory of data')
```

```python
currentframe = 0
```

```python
if (cap.isOpened()== False):  
    print("Error opening video  file") 
   
```

```python
#Converting videos into frames
```

```python
while(cap.isOpened()):
    cap.set(cv2.CAP_PROP_POS_MSEC,(currentframe*1000))
    ret, frame = cap.read()
    if ret == True:
        name = './data_las/frame' + str(round(currentframe,2)) + '.jpg'
        print ('Creating...' + name)
        cv2.imwrite(name, frame)
        currentframe += 0.1
    else:
        break
        
cap.release() 
cv2.destroyAllWindows()
        
    
```

```python
#importind all the frames from the folder created
```

```python
images = []
for frames in os.listdir(r"C:\Users\dhananjayan\project_submission\data_1"):
    img = cv2.imread(os.path.join(r"C:\Users\dhananjayan\project_submission\data_1",frames))
    if img is not None:
        images.append(img)
print(images)
    
```

```python
#importing the created module for stitching
```

```python
from panorama import Panaroma
import imutils
```

```python
#reducing the length of the list with frames cause the truch was present in the video less than a half the length of video
```

```python
no_of_images =len(images)
```

```python
#resizing the images
```

```python
for i in range(no_of_images):
    images[i] = imutils.resize(images[i], width=400)
```

```python
for i in range(no_of_images):
    images[i] = imutils.resize(images[i], height=400)
```

```python
#stiching the given number of images
```

```python
panaroma = Panaroma()
if no_of_images==2:
    (result, matched_points) = panaroma.image_stitch([images[0], images[1]], match_status=True)
else:
    (result, matched_points) = panaroma.image_stitch([images[no_of_images-2], images[no_of_images-1]], match_status=True)
    for i in range(no_of_images - 2):
        (result, matched_points) = panaroma.image_stitch([images[no_of_images-i-3],result], match_status=True)
```

```python
#saving the panoromic image
```

```python
cv2.imshow("Keypoint Matches", matched_points)
cv2.imshow("Panorama", result)
```

```python
cv2.imwrite("Matched_points.jpg",matched_points)
cv2.imwrite("Panorama_image.jpg",result)
```
